```python
data_2016 = pd.read_csv('/Users/Anna/Documents/GitHub/Speciale - Python/pixel_values/pixel_values_2016.csv', index_col=0)
data_2016 = data_2016.rename(columns={'ID':'ID','layer':'2016'})
```

```python
data_2017 = pd.read_csv('/Users/Anna/Documents/GitHub/Speciale - Python/pixel_values/pixel_values_2017.csv', index_col=0)
data_2017 = data_2017.rename(columns={'ID':'ID','layer':'2017'})
```

```python
data_2018 = pd.read_csv('/Users/Anna/Documents/GitHub/Speciale - Python/pixel_values/pixel_values_2018.csv', index_col=0)
data_2018 = data_2018.rename(columns={'ID':'ID','layer':'2018'})
```

```python
data_2019 = pd.read_csv('/Users/Anna/Documents/GitHub/Speciale - Python/pixel_values/pixel_values_2019.csv', index_col=0)
data_2019 = data_2019.rename(columns={'ID':'ID','layer':'2019'})
```

```python
pixe_values_1918 = pd.merge(data_2019, data_2018[['cell','2018']], on='cell')
pixe_values_1817 = pd.merge(data_2018, data_2017[['cell','2017']], on='cell')
pixe_values_1716 = pd.merge(data_2017, data_2016[['cell','2016']], on='cell')
pixe_values_1615 = pd.merge(data_2016, data_2015[['cell','2015']], on='cell')
pixe_values_1514 = pd.merge(data_2015, data_2014[['cell','2014']], on='cell')
```

```python
pixe_values_1918['TH_1'] = pixe_values_1918['2019']-pixe_values_1918['2018']>300
pixe_values_1918['TH_2'] = pixe_values_1918['2019']>1150
pixe_values_1918['CHANGE 1918'] = pixe_values_1918['TH_1']*pixe_values_1918['TH_2']

pixe_values_1817['TH_1'] = pixe_values_1817['2018']-pixe_values_1817['2017']>300
pixe_values_1817['TH_2'] = pixe_values_1817['2018']>1150
pixe_values_1817['CHANGE 1817'] = pixe_values_1817['TH_1']*pixe_values_1817['TH_2']

pixe_values_1716['TH_1'] = pixe_values_1716['2017']-pixe_values_1716['2016']>300
pixe_values_1716['TH_2'] = pixe_values_1716['2017']>1150
pixe_values_1716['CHANGE 1716'] = pixe_values_1716['TH_1']*pixe_values_1716['TH_2']

pixe_values_1615['TH_1'] = pixe_values_1615['2016']-pixe_values_1615['2015']>300
pixe_values_1615['TH_2'] = pixe_values_1615['2016']>1150
pixe_values_1615['CHANGE 1615'] = pixe_values_1615['TH_1']*pixe_values_1615['TH_2']

pixe_values_1514['TH_1'] = pixe_values_1514['2015']-pixe_values_1514['2014']>300
pixe_values_1514['TH_2'] = pixe_values_1514['2015']>1150
pixe_values_1514['CHANGE 1514'] = pixe_values_1514['TH_1']*pixe_values_1514['TH_2']
```

```python
Mine_changes = pd.DataFrame(pixe_values_1918.groupby('ID')['CHANGE 1918'].sum())
Mine_changes = pd.merge(Mine_changes, pd.DataFrame(pixe_values_1817.groupby('ID')['CHANGE 1817'].sum()), how='left', on='ID')
Mine_changes = pd.merge(Mine_changes, pd.DataFrame(pixe_values_1716.groupby('ID')['CHANGE 1716'].sum()), how='left', on='ID')
Mine_changes = pd.merge(Mine_changes, pd.DataFrame(pixe_values_1615.groupby('ID')['CHANGE 1615'].sum()), how='left', on='ID')
Mine_changes = pd.merge(Mine_changes, pd.DataFrame(pixe_values_1514.groupby('ID')['CHANGE 1514'].sum()), how='left', on='ID')
```

```python
Mine_changes = Mine_changes.fillna(0)
```
